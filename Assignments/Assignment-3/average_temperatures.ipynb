```python
import numpy as np
import pylab as pl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#code to load files 
list_of_files = [('montreal_temperature_2012.dat','Montreal'),
                 ('vancouver_temperature_2012.dat','Vancouver'),
                 ('toronto_temperature_2012.dat','Toronto')]

datalist = [(pl.loadtxt(filename), label) for filename, label in list_of_files]

for data, label in datalist:
    pl.plot(data, label=label)
    
pl.title('2012 Average Monthly Temperatures')
pl.xlabel('months(2012)')
pl.ylabel('avg monthly temperature (celcius)')
plt.legend(loc='lower center', frameon=True)
plt.show()
```
